### Replace codes with real values

#### I was using document CodeDescriptions.pdf (link already provided above in Comment markdown cell of part "2. Reading data") to replace codes in several columns with appropriate values.
#### However, I will not change codes in column 'utilities' because they represent combination of first letters of available utilities at property. W - water, S - sewer, G - gas, E - electric or ALL. So, SGE is sewer, gas, electric. This would take too much space.
#### Columns that will have codes changed are: 'story_height', 'design_style', 'bath' and 'type_and_use'. Codes in 'story_height' were converted to decimal numbers.

```python
# Replacing codes in column 'story_height'
df_residential = df_residential.replace({'story_height':{'A':'1 story', 'B':'1.5 story', 'C':'2 story', 'D':'2.5 story', 'E':'3 story', 'F':'3.5 story', 'G':'4 story', 'H':'Multi story', 'I':'1.75 story', 'J':'1.4 story', 'K':'1.63 story', 'L':'1.88 story', 'M':'2.4 story', 'N':'2.63 story', 'O':'2.75 story'}})
```

```python
df_residential['story_height'].value_counts(dropna = False)
```

```python
# For now, I'll change NaN to Unknown
df_residential['story_height'] = df_residential['story_height'].fillna('Unknown')
```

```python
df_residential['story_height'].value_counts(dropna = False)
```

```python
# replacing codes in column 'design_style'
df_residential = df_residential.replace({'design_style':{'A':'Conventional', 'B':'Duplex', 'C':'Townhouse', 'D':'Condo', 'E':'Conversion', 'F':'Colonial', 'G':'Ranch', 'H':'Cape', 'I':'Split level', 'J':'Split foyer', 'K':'Contemporary', 'L':'Log', 'M':'Manuf sngl', 'N':'Manuf multi', 'O':'Modular'}})
```

```python
# For now, I'll change NaN and P to Unknown
df_residential['design_style'] = df_residential['design_style'].fillna('Unknown')
```

```python
df_residential = df_residential.replace({'design_style':{'P':'Unknown'}})
```

```python
df_residential['design_style'].value_counts(dropna = False)
```

```python
# Replacing codes in column 'bath'
df_residential = df_residential.replace({'bath':{'A':'1 bath', 'B':'1.5 bath', 'C':'2 bath', 'D':'2.5 bath', 'E':'3 bath', 'F':'3.5 bath', 'G':'Limited plmg', 'H':'No plumbing', 'I':'Adequate', 'J':'NO of fixtures'}})
```
